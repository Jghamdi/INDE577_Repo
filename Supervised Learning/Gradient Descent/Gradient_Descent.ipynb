# Gradient Descent

In this notebook, we'll explore the concept of gradient descent through a simple example of linear regression. Gradient descent is a fundamental optimization algorithm used to minimize the loss function during the training of machine learning models.

We'll start by explaining the intuition behind gradient descent and then demonstrate its application in optimizing the parameters of a linear regression model.

---


#### Step 1: Generate Synthetic Data

To illustrate gradient descent, we'll create synthetic data representing a simple linear relationship between a feature \(X\) and a target variable \(y\). We'll generate 100 data points with a known linear equation \(y = 4 + 3X\) plus some random noise.

Let's move on to initializing the parameters and defining the loss function.

```python
# Import necessary libraries for the code.
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data.
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

#### Steps 2-4: Initialize Parameters, Define Loss Function, and Compute Gradients


Now that we have our data, let's move on to setting up our linear regression model. We'll initialize the slope and intercept parameters and define the mean squared error (MSE) loss function. Additionally, we'll define a function to compute the gradients of the loss function with respect to the parameters.

Next, we'll proceed to update the parameters using gradient descent.

```python
# Step 2: Initialize parameters.
m = 0  # initial slope
b = 0  # initial intercept

# Step 3: Define the loss function (Mean Squared Error).
def mse_loss(y_true, y_pred):
    """
    Calculate the Mean Squared Error (MSE) loss between the true values and the predicted values.
    
    Parameters:
        y_true (array-like): The true target values.
        y_pred (array-like): The predicted target values.
        
    Returns:
        float: The mean squared error (MSE) loss between the true and predicted values.
    """
    return np.mean((y_true - y_pred) ** 2)

# Step 4: Define the gradient function.
def compute_gradients(X, y, y_pred):
    """
    Compute the gradients of the loss function with respect to the slope and intercept parameters.
    
    Parameters:
        X (array-like): The feature values.
        y (array-like): The true target values.
        y_pred (array-like): The predicted target values.
        
    Returns:
        tuple: A tuple containing the gradients of the loss function with respect to the slope and intercept parameters.
    """
    d_m = -2 * np.mean(X * (y - y_pred))  # Gradient of loss w.r.t. slope
    d_b = -2 * np.mean(y - y_pred)       # Gradient of loss w.r.t. intercept
    return d_m, d_b
```

#### Steps 5: Update Parameters using Gradient Descent

Now comes the core of the gradient descent algorithm. We'll update the parameters (slope and intercept) iteratively using the computed gradients in the opposite direction to minimize the loss function.

As we iterate through the training process, we'll keep track of the loss values to visualize how the loss decreases over time.

Let's proceed to visualize the results.


```python
# Define the learning rate and the number of iterations.
learning_rate = 0.01
num_iterations = 100

# Create an empty list to store the loss values over iterations.
losses = []

# Iterate over the specified number of iterations.
for _ in range(num_iterations):
    # Compute the predictions using the current parameters.
    y_pred = m * X + b
    
    # Calculate the mean squared error loss between the true and predicted values.
    loss = mse_loss(y, y_pred)
    
    # Append the loss value to the list of losses.
    losses.append(loss)
    
    # Compute the gradients of the loss function with respect to the parameters.
    d_m, d_b = compute_gradients(X, y, y_pred)
    
    # Update the parameters using gradient descent.
    m -= learning_rate * d_m
    b -= learning_rate * d_b
```

#### Step 6: Visualize the Results

After training the linear regression model using gradient descent, we can now visualize the data points along with the best-fitting line obtained by the optimized parameters.

Let's move on to plotting the loss over iterations to analyze the convergence of our optimization process.


```python
# Step 6: Plot the data and the best-fitting line.
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Data points')
plt.plot(X, m * X + b, color='red', label='Best-fitting line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression with Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```

#### Visualizing Loss over Iterations

Finally, we plot the loss function over iterations to observe how it decreases over time. This plot provides insights into the convergence behavior of our optimization process.

That concludes our demonstration of gradient descent in the context of linear regression. We hope this notebook helps you understand the fundamental concepts behind gradient descent and its application in machine learning optimization.

Feel free to experiment with different hyperparameters or extend this example to more complex models!


```python
# Plot the loss over iterations.
plt.figure(figsize=(10, 6))
plt.plot(range(num_iterations), losses)
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Loss over Iterations')
plt.grid(True)
plt.show()
```
